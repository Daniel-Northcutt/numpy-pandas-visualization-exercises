```python
import math

#x = list(range(0,50))

y1 = [((0*n)+1) for n in x]
y2 = np.log(x)
y3 = x
y4 = x * np.log(x)
y5 = [(n**2) for n in x]
y6 = [(2**n) for n in x]
y7 = [(math.factorial(n)) for n in x]
y8 =[(n **n) for n in x]

plt.plot(x,y1, label = 'O(1)')
plt.plot(x,y2, label = 'O(logn)')
plt.plot(x,y3, label = 'O = n')
plt.plot(x,y4, label = 'O(n log n)')
plt.plot(x,y5, label = 'O(n^)')
plt.plot(x,y6, label = 'O(2^n)')
plt.plot(x,y7, label = 'O(n!)')
plt.plot(x,y8, label = 'O(n^n)')
plt.annotate("Daniel Northcutt", (8, 13))

plt.xlim(0,15)
plt.ylim(0,15)
plt.legend()
plt.title('Big O Notation')
plt.xlabel ('Elements')
plt.ylabel ('Operations')
plt.tight_layout()

plt.show()
```
