```python
#1 Plot the following equation y=((x**2) x + 2). You'll need to write the code that generates the x and y points.
#Add an anotation for the point 0, 0, the orig

from matplotlib import pyplot as plt
import numpy as np


x = list(range(-100,100)) #define x
y = [((x**2) - x + 2) for x in x] # y is equation to plot
plt.annotate('(0,0) - Origin', xy=(0,0), xytext=(60,10),
             arrowprops={'facecolor' : 'black'})

plt.xlabel('Range')
plt.ylabel('y=x^2 - (x +2)$')
plt.title('Plot 1');

plt.plot(x, y);
```

```python
plt.xkcd()

x = list(range(0,50))

y1 = [math.sqrt(n) for n in x]
y2 = [(n**3) for n in x]
y3 = ([2**n for n in x])
y4 = [1/(n+1) for n in x]

plt.plot(x, y1, color = 'red', label = '$sqrt(x)$')
plt.plot(x, y2, color = 'blue', label = '$x^3$')
plt.plot(x, y3, color = 'green', label = '$2^x$')         
plt.plot(x, y4, color = 'black', label = '$1/(x+1)$')
plt.xlim(0,10)
plt.ylim(0,10)
plt.title('Math Functions')
plt.xlabel ('$x$')
plt.ylabel ('$y$')
plt.legend()
plt.show()
```
